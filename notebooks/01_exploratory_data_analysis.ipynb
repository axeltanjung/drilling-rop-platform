# Drilling Telemetry — Exploratory Data Analysis

EDA for the Oil & Gas Drilling ROP Prediction & Optimization Platform.
Run the data generator first: `python backend/data/synthetic_drilling_data_generator.py`

```python
import sys, pathlib
sys.path.append(str(pathlib.Path.cwd().parent))
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
from backend.utils.config import settings

df = pd.read_csv(settings.raw_data_path, parse_dates=['timestamp'])
print(df.shape)
df.head()
```

```python
df.describe().T
```

## ROP by formation (physics check: hard formations drill slower)

```python
rop_by_form = df.groupby('formation_type')['rate_of_penetration'].mean().sort_values()
rop_by_form.plot(kind='barh', color='#22d3ee')
plt.title('Average ROP by Formation'); plt.xlabel('ROP (ft/hr)'); plt.show()
```

## Correlation with ROP

```python
num = df.select_dtypes('number')
corr = num.corr()['rate_of_penetration'].sort_values(ascending=False)
corr
```

## WOB vs ROP scatter (diminishing returns + vibration penalty)

```python
s = df.sample(5000)
plt.scatter(s['weight_on_bit'], s['rate_of_penetration'], c=s['vibration_level'], cmap='plasma', s=6)
plt.colorbar(label='Vibration'); plt.xlabel('WOB (klbs)'); plt.ylabel('ROP (ft/hr)'); plt.show()
```

## Bit wear vs efficiency

```python
s = df.sample(5000)
plt.scatter(s['bit_wear'], s['drilling_efficiency_score'], s=6, color='#f59e0b', alpha=0.4)
plt.xlabel('Bit Wear'); plt.ylabel('Efficiency Score'); plt.show()
```
